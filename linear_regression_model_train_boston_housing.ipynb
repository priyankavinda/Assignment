# to train a model
# 1 step  load the librery

```python
from sklearn.linear_model import LinearRegression   # select the model
from sklearn.model_selection import train_test_split, cross_val_score   # training and validation
from sklearn.metrics import mean_squared_error, r2_score    # performance of the model
```

## 2ed Step split the data

```python
# Split the data for trining and validation as 80% and 20%
x_train_1, x_val_1, y_train_1, y_val_1 = train_test_split(x,y, test_size=0.2, random_state=5)
```

# train the model

```python
slrm = LinearRegression()
slrm.fit(x_train_1,y_train_1)
```

```python
y_predicted_1 = slrm.predict(x_val_1)
```

```python
y_predicted_1.shape
```

## measure the model performance

```python
mean_squared_error = mean_squared_error(y_predicted_1, y_val_1)
```

```python
mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error)
```

```python
rmse
```

```python
y.mean()
```

```python
error = rmse/y.mean()*100
print('Error:',error)
print('Accuracy:',100-error )
```
